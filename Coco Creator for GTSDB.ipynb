```python
#gt.txt into ann
ann = data[['img', 'x1', 'y1', 'x2', 'y2', 'id']]
ann
```

```python
ann['img'] = ann['img'].str.replace(r'.ppm$', '.jpg')
ann
```

```python
#locate values from gt.txt which are in the train/test folder as images 
#replace test and train in the isin list 
only_train = ann[ann['img'].isin(test)]
print(len(only_train))
only_train
```

```python
#replace the for loop only train/test
iscrowd = 0
with open("D:\GTSDB\Datasets\FullIJCNN2013\Coco/test_annotations_key.txt", "w") as text_file:
        text_file.write('"annotations":[\n')
        for index, rows in only_train.iterrows():
            x1 = rows['x1']
            y1 = rows['y1']
            width = abs(rows['x2'] - rows['x1'])
            height = abs(rows['y2'] -rows['y1'])
            area = width * height
            text_file.write('{"id":%s,' % str(index))
            text_file.write('"category_id":%s,' % rows['id'])
            text_file.write('"iscrowd":%s,' % str(iscrowd))
            text_file.write('"segmentation": [[]],')
            text_file.write('"image_id":%d,' % int(rows['img'][:-4]))
            text_file.write('"area":%s,' % str(area))
            text_file.write('"bbox":[%s,' % str(x1))
            text_file.write('%s,' % str(y1))
            text_file.write('%s,' % str(width))
            text_file.write('%s]},\n' % str(height))
        text_file.write(']')
```
